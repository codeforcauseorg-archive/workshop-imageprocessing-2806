```python
import cv2

cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    if ret:
        cv2.imshow("My Camera", frame)

    key = cv2.waitKey(10)

    if key == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
